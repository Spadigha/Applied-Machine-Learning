# Q-Q Plot

```python
#Q-Q plot
import numpy as np 
import pylab 
import scipy.stats as stats

# N(0,1)
# loc = Mean and scale = std-dev
std_normal = np.random.normal(loc = 0, scale = 1, size=1000)

# 0 to 100th percentiles of std-normal
for i in range(0,101):
    print("{} Percentile : {}".format(i, np.percentile(std_normal,i)))

```

```python
# generate 100 sanples from N(20,5)
measurements = np.random.normal(loc = 20, scale = 5, size=100) 
#try size=1000 or greater; More points will be nearer to line 


stats.probplot(measurements, dist="norm", plot=pylab)
pylab.show()

# Notte: Descrepancy  can be due to small sample size
```

```python
# 5000 obsvns
measurements = np.random.normal(loc = 20, scale = 5, size=5000)
stats.probplot(measurements, dist="norm", plot=pylab)
pylab.show()

# Note: More guassian as the sample size increased!
```

##### Disadvantage:  Q-Q PLot won't work if number of samples/obsvns are `less`. It will become hard to interpret it.

```python
# generate 100 sanples from uniform distrbn.
measurements = np.random.uniform(low=-1, high=1, size=100) 
#try size=1000

stats.probplot(measurements, dist="norm", plot=pylab)
pylab.show()

# Note: Hard to differentiate between uniform or guassian when sample size is less.
# Hence, the disadvantage
```

```python
measurements = np.random.uniform(low=-1, high=1, size=10000) 
stats.probplot(measurements, dist="norm", plot=pylab)
pylab.show()

# Note: Easy to   ddifferentiate when samples are more!
```
