# K-S Test

```python
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt

#generate a gaussian r.v X
x = stats.norm.rvs(size=1000);
sns.set_style('whitegrid')
sns.kdeplot(np.array(x), bw=0.5)
plt.show()


```

```python
stats.kstest(x, 'norm')
```

```python
# Y ~ Continous Uniform Distribution(0,1)
y = np.random.uniform(0,1,10000);
sns.kdeplot(np.array(y), bw=0.1)
plt.show()
```

```python
# Note: This is uniform distribution. Yet, we see a curve in begining and end simply because of estimation of KDE. 
#       It actually starts from 0 and 1 as shown below. Not shown above as from -0.2 to 1.2 

plt.hist(y)
plt.show()
```

```python
stats.kstest(y, 'norm')
```
